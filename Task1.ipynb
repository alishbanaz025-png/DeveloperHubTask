```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# 1. Load the dataset
iris_data = load_iris()
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris.columns = [c.replace(' (cm)', '').replace(' ', '_') for c in iris.columns]
iris['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

# 2. Display dataset structure
print("Shape:", iris.shape)
print("Columns:", iris.columns)
print(iris.head())

# 3. Visualizations
# Scatter plot
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.savefig('task1_scatter.png')
plt.clf()

# Histogram
sns.histplot(data=iris, x='petal_length', kde=True)
plt.title('Petal Length Distribution')
plt.savefig('task1_histogram.png')
plt.clf()

# Box plot
sns.boxplot(data=iris, x='species', y='petal_width', hue='species', palette='Set2')
plt.title('Petal Width by Species')
plt.savefig('task1_box.png')
plt.clf()
```
